```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
nums = []
for i in range(100):
    nums.append(random.uniform(0, 200))
```

```python
def teta1(n):
    return abs(2*np.sum(nums[:n])/n - 200)
```

```python
def teta2(n):
    return abs((n+1)*np.amin(nums[:n]) - 200)
```

```python
def teta3(n):
    return abs((1+1/n)*np.amax(nums[:n]) - 200)
```

```python
def teta4(n):
    return abs(np.amin(nums[:n]) + np.amax(nums[:n]) - 200)
```

```python
def teta5(n):
    return abs(nums[0] + np.sum(nums[1:n])/(n-1) - 200)
```

```python
y1 = [teta1(n) for n in range(99)]
x = [i for i in range(99)]
y2 = [teta2(n) for n in range(1, 100)]
y3 = [teta3(n) for n in range(1, 100)]
y4 = [teta4(n) for n in range(1, 100)]
y5 = [teta5(n) for n in range(99)]
plt.figure(figsize=(10,6))
plt.plot(x, y1, label = 'teta1')
plt.plot(x, y2, label = 'teta2')
plt.plot(x, y3, label = 'teta3')
plt.plot(x, y4, label = 'teta4')
plt.plot(x, y5, label = 'teta5')
plt.title('Task1')
plt.xlabel('n')
plt.ylabel('|teta - ~teta|')
plt.legend()
plt.savefig('task1.png')
plt.show()
```
